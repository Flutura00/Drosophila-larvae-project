```python
plt.hist(df_round_dist['angle_pointing'])
plt.title("Histogram of angle pointing in all the data")
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (7, 7)
bins_count,pdf,cdf =  pdf_cdf(lis_val = df_round_dist['angle_pointing'],from_val = 0,to_val = 360,bins=5)
plt.plot(bins_count[1:],pdf,label = "all")
plt.title("bla")
plt.legend()
plt.title("All data angles of pointing, probability distribution function")
plt.show()
```

```python
plt.show()
plt.rcParams["figure.figsize"] = (7, 7)
colors = ['red', 'blue', 'yellow', 'green', 'magenta']
for stim in range(0, len(stimuli)):
    df_st = df_round_dist[df_round_dist["stimulus_name"]==stimuli[stim]]
    arr = df_st['angle_pointing'].values
    bla = arr[~np.isnan(arr)]
    circl_plot(angles = bla)
    plt.title(f"probability distribution of angles in stimuli {stimuli[stim]}")
    plt.legend()
    plt.show()
```

```python
plt.rcParams["figure.figsize"] = (7, 7)
colors = ['red','blue','yellow','green','cyan',"purple","orange","teal","cyan"]
for stim in range(0,len(stimuli)):
    df_st = df_round_dist[df_round_dist["stimulus_name"]==stimuli[stim]]
    bins_count,pdf,cdf =  pdf_cdf(lis_val = df_st['angle_pointing'],from_val = -150,to_val = 150,bins=30)
    plt.plot(bins_count[1:],pdf,label = stimuli[stim])
plt.title("probability distribution of angles in different stimuli",size =18)
plt.legend()
plt.show()
```
